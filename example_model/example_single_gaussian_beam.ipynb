# Распространение одиночного гауссова пучка

Рассматривается скалярное комплексное поле гауссова пучка в поперечной плоскости. Свободное распространение вычисляется методом углового спектра.

```python
import numpy as np
import matplotlib.pyplot as plt

from AO import GaussianBeam, Propagator
```

## Параметры расчёта

Размер расчётной области выбирается так, чтобы интенсивность пучка была локализована внутри численной апертуры.

```python
N = 1024
size_screen = 0.05
wavelength = 650e-9
distance = 50.0
w0 = 0.002

dx = size_screen / N
extent = [-size_screen / 2, size_screen / 2, -size_screen / 2, size_screen / 2]
```

## Начальное поле

Начальное поле задаётся гауссовым распределением амплитуды. Фаза в начальной плоскости считается постоянной.

```python
beam = GaussianBeam(
    N=N,
    size_screen=size_screen,
    w0=w0
)

E0 = beam.get_field()
I0 = beam.get_intensity(normalized=True)
phi0 = beam.get_phase()
```

```python
plt.figure(figsize=(7, 6))
image = plt.imshow(I0, origin="lower", extent=extent, cmap = 'hot')
plt.xlabel("x, м")
plt.ylabel("y, м")
plt.colorbar(image, label="I / Imax")
plt.show()
```

```python
plt.figure(figsize=(7, 6))
image = plt.imshow(phi0, origin="lower", extent=extent)
plt.xlabel("x, м")
plt.ylabel("y, м")
plt.colorbar(image, label="arg E, рад")
plt.show()
```

## Свободное распространение

Поле в выходной плоскости получается умножением пространственного спектра на передаточную функцию свободного пространства.

```python
E1 = Propagator.propagate(
    N=N,
    dx=dx,
    E0=E0,
    wavelength=wavelength,
    distance=distance,
    method="exact"
)

I1 = np.abs(E1) ** 2
I1 = I1 / I1.max()
phi1 = np.angle(E1)
```

```python
plt.figure(figsize=(7, 6))
image = plt.imshow(I1, origin="lower", extent=extent, cmap = 'hot')
plt.xlabel("x, м")
plt.ylabel("y, м")
plt.colorbar(image, label="I / Imax")
plt.show()
```

```python
plt.figure(figsize=(7, 6))
image = plt.imshow(phi1, origin="lower", extent=extent)
plt.xlabel("x, м")
plt.ylabel("y, м")
plt.colorbar(image, label="arg E, рад")
plt.show()
```

## Контроль мощности

Для свободного распространения без апертурных ограничений интегральная мощность должна сохраняться с точностью численной схемы.

```python
P0 = np.sum(np.abs(E0) ** 2) * dx**2
P1 = np.sum(np.abs(E1) ** 2) * dx**2
relative_error = abs(P1 - P0) / P0

print(f"P0 = {P0:.6e}")
print(f"P1 = {P1:.6e}")
print(f"Относительная ошибка = {relative_error:.3e}")
```
